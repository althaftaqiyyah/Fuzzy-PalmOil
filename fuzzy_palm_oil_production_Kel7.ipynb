```python
import matplotlib
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

## Sterilising and Threshing

```python
#Deklarasi input variables
X12 = np.arange(0, 18, 1)
X22 = np.arange(70, 150, 1)
X32 = np.arange(2.5, 3.3, 0.01)

#Deklarasi output variables
OW12 = np.arange(0, 0.6, 0.01)
OW13 = np.arange(0, 4, 0.01)
OW23 = np.arange(0, 1.2, 0.01)
```

```python
#deklarasi membership function input
X12_low = fuzz.trapmf(X12, [0, 0, 3, 9])
X12_high = 1-X12_low

X22_short = fuzz.trapmf(X22, [70, 70, 90, 110])
X22_medium = fuzz.trimf(X22, [90, 110, 130])
X22_long = fuzz.trapmf(X22, [110, 130, 150, 150])

X32_low = fuzz.trapmf(X32, [2.5, 2.5, 2.7, 2.9])
X32_medium = fuzz.trimf(X32, [2.7, 2.9, 3.1])
X32_high = fuzz.trapmf(X32, [2.9, 3.1, 3.3, 3.3])
```

```python
#deklarasi membership function output
OW12_low = fuzz.trapmf(OW12, [0, 0, 0.2, 0.3])
OW12_medium = fuzz.trimf(OW12, [0.2, 0.3, 0.4])
OW12_high = fuzz.trapmf(OW12, [0.3, 0.4, 0.6, 0.6])

OW13_low = fuzz.trapmf(OW13, [0, 0, 1.5, 2])
OW13_medium = fuzz.trimf(OW13, [1.5, 2, 2.5])
OW13_high = fuzz.trapmf(OW13, [2, 2.5, 4, 4])

OW23_low = fuzz.trapmf(OW23, [0, 0, 0.2, 0.35])
OW23_medium = fuzz.trimf(OW23, [0.2, 0.35, 0.5])
OW23_high = fuzz.trapmf(OW23, [0.35, 0.5, 1.25, 1.25])
```

```python
#visualisasi membership function fuzzy set
fig, (ax0, ax1, ax2, ax3, ax4, ax5) = plt.subplots(nrows = 6, figsize=(15,16))

ax0.plot(X12, X12_low, 'r', linewidth=1.5, label='low')
ax0.plot(X12, X12_high, 'b', linewidth=1.5, label='high')
ax0.set_title('Membership function fuzzy set var X12')
ax0.legend()

ax1.plot(X22, X22_short, 'r', linewidth=1.5, label='short')
ax1.plot(X22, X22_medium, 'b', linewidth=1.5, label='medium')
ax1.plot(X22, X22_long, 'g', linewidth=1.5, label='long')
ax1.set_title('Membership function fuzzy set var X22')
ax1.legend()

ax2.plot(X32, X32_low, 'r', linewidth=1.5, label='low')
ax2.plot(X32, X32_medium, 'b', linewidth=1.5, label='medium')
ax2.plot(X32, X32_high, 'g', linewidth=1.5, label='high')
ax2.set_title('Membership function fuzzy set var X32')
ax2.legend()

ax3.plot(OW12, OW12_low, 'r', linewidth=1.5, label='low')
ax3.plot(OW12, OW12_medium, 'b', linewidth=1.5, label='medium')
ax3.plot(OW12, OW12_high, 'g', linewidth=1.5, label='high')
ax3.set_title('Membership function fuzzy set var OW12')
ax3.legend()

ax4.plot(OW13, OW13_low, 'r', linewidth=1.5, label='low')
ax4.plot(OW13, OW13_medium, 'b', linewidth=1.5, label='medium')
ax4.plot(OW13, OW13_high, 'g', linewidth=1.5, label='high')
ax4.set_title('Membership function fuzzy set var OW13')
ax4.legend()

ax5.plot(OW23, OW23_low, 'r', linewidth=1.5, label='low')
ax5.plot(OW23, OW23_medium, 'b', linewidth=1.5, label='medium')
ax5.plot(OW23, OW23_high, 'g', linewidth=1.5, label='high')
ax5.set_title('Membership function fuzzy set var OW23')
ax5.legend()
```

```python
#fuzzifikasi variabel input X12
x1 = 6

x12_lo_degree = fuzz.interp_membership(X12, X12_low, x1)
x12_hi_degree = fuzz.interp_membership(X12, X12_high, x1)

x12_fuzz_var = np.array([x12_lo_degree, x12_hi_degree])
print('Fuzzy Variable(X12 =' , x1, ') :', x12_fuzz_var)

#fuzzifikasi variabel input x22
y1 = 110

x22_sh_degree = fuzz.interp_membership(X22, X22_short, y1)
x22_md_degree = fuzz.interp_membership(X22, X22_medium, y1)
x22_lg_degree = fuzz.interp_membership(X22, X22_long, y1)

x22_fuzz_var = np.array([x22_sh_degree, x22_md_degree, x22_lg_degree])
print('Fuzzy Variable(X22 =' , y1, ') :', x22_fuzz_var)

#fuzifikasi variabel input x32
z1 = 2.9

x32_lo_degree = fuzz.interp_membership(X32, X32_low, z1)
x32_md_degree = fuzz.interp_membership(X32, X32_medium, z1)
x32_hi_degree = fuzz.interp_membership(X32, X32_high, z1)

x32_fuzz_var = np.array([x32_lo_degree, x32_md_degree, x32_hi_degree])
print('Fuzzy Variable(X32 =' , z1, ') :', x32_fuzz_var)
```
